### Compairing two output neurons to one output neuron. 
----
LR:         0.01 \
epochs:     50 \
batch size: 50 \
optimizer: SGD

```python
# Read data
validation_one_neuron = pd.read_csv('/data/khood/GitHub/MLAudio/run2/training_one_neuron.csv')
validation_two_neuron = pd.read_csv('/data/khood/GitHub/MLAudio/run2/training_two_neuron.csv')

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot for one output neuron
axs[0, 0].plot(validation_one_neuron['losses'], label='Losses')
axs[0, 0].set_title('Training Losses over Iterations \n one output neuron')
axs[0, 0].set_xlabel('Iterations')
axs[0, 0].set_ylabel('Loss')
axs[0, 0].legend()

axs[0, 1].plot(validation_one_neuron['accuracies'], label='Accuracies', color='green')
axs[0, 1].set_title('Training Accuracies over Iterations \n one output neuron')
axs[0, 1].set_xlabel('Iterations')
axs[0, 1].set_ylabel('Accuracy')
axs[0, 1].legend()

# Plot for two output neurons
axs[1, 0].plot(validation_two_neuron['losses'], label='Losses')
axs[1, 0].set_title('Training Losses over Iterations \n two output neuron')
axs[1, 0].set_xlabel('Iterations')
axs[1, 0].set_ylabel('Loss')
axs[1, 0].legend()

axs[1, 1].plot(validation_two_neuron['accuracies'], label='Accuracies', color='green')
axs[1, 1].set_title('Training Accuracies over Iterations \n two output neuron')
axs[1, 1].set_xlabel('Iterations')
axs[1, 1].set_ylabel('Accuracy')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot for one output neuron
axs[0, 0].plot(validation_one_neuron[35000:]['losses'], label='Losses')
axs[0, 0].set_title('Training Losses over Iterations \n one output neuron')
axs[0, 0].set_xlabel('Iterations')
axs[0, 0].set_ylabel('Loss')
axs[0, 0].legend()

axs[0, 1].plot(validation_one_neuron[35000:]['accuracies'], label='Accuracies', color='green')
axs[0, 1].set_title('Training Accuracies over Iterations \n one output neuron')
axs[0, 1].set_xlabel('Iterations')
axs[0, 1].set_ylabel('Accuracy')
axs[0, 1].legend()

# Plot for two output neurons
axs[1, 0].plot(validation_two_neuron[35000:]['losses'], label='Losses')
axs[1, 0].set_title('Training Losses over Iterations \n two output neuron')
axs[1, 0].set_xlabel('Iterations')
axs[1, 0].set_ylabel('Loss')
axs[1, 0].legend()

axs[1, 1].plot(validation_two_neuron[35000:]['accuracies'], label='Accuracies', color='green')
axs[1, 1].set_title('Training Accuracies over Iterations \n two output neuron')
axs[1, 1].set_xlabel('Iterations')
axs[1, 1].set_ylabel('Accuracy')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python
print(f"accuracies variance {math.sqrt(validation_one_neuron[35000:].std()['accuracies'])}") 
validation_one_neuron[35000:].mean()
```

```python
print(f"training accuracies variance multi class {math.sqrt(validation_two_neuron[35000:].std()['accuracies'])}")
validation_two_neuron[35000:].mean()
```
