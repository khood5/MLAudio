# Visualization for log outputs of SNNTorch training runs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('d.txt', 'r') as f:
    data = [[float(entry.replace('\n', '')) for entry in line.split(',')] for line in f.readlines()]

data = np.array(data)
```

```python
plt.plot(np.linspace(0, len(data), len(data)), data[:, 0], label='Train Loss')
plt.plot(np.linspace(0, len(data), len(data)), data[:, 1], label='Validation Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
plt.title('Validation set Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy %')
plt.plot(np.linspace(0, len(data), len(data)), data[:, 2])
```
