# Visualization for log outputs of SNNTorch training runs

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_log(path):
    with open(path, 'r') as f:
        data = [[float(entry.replace('\n', '')) for entry in line.split(',')] for line in f.readlines()]
    
    return np.array(data)
```

```python
# load everything
all_data = []

RUN_COUNT = 9
BASE_PATH = './results/'

for i in range(RUN_COUNT):
    l = load_log(BASE_PATH+str(i)+'/log.csv')
    all_data.append(l)
```

# Some runs on various datasets
All of these use model size INPUTx100x100x2, LR=0.005 and batch size 128, with beta decay value 0.9

### 5k Samples FFT spectrogram using Pos-Neg Delta Modulation
Normalized globally (using min and max from all samples), DM threshold is 0.001

```python
INDEX = 1

plt.plot(np.linspace(0, len(all_data[INDEX]), len(all_data[INDEX])), all_data[INDEX][:, 0], label='Train Loss')
plt.plot(np.linspace(0, len(all_data[INDEX]), len(all_data[INDEX])), all_data[INDEX][:, 1], label='Validation Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss')
```

```python
plt.title('Validation set Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy %')
plt.plot(np.linspace(0, len(all_data[INDEX]), len(all_data[INDEX])), all_data[INDEX][:, 2], color='purple')
```

### 5k FFT spectrogram using Pos-Neg Delta Modulation (smaller dm threshold)
Normalized **globally** (using min and max out of all samples), DM threshold is **0.0001**

Picked best model out of all from before, reduced delta mod threshold. Reasoning is that with the 0.001 threshold, about 15% of our samples were spking less than 10 times total (because their intensity values in the spectrogram are very small, since we normalized globally). After lowering delta mod, this number reduces to 4%.
