```python
# For retrieving data from console output of train_script.py 
# I forgot to save the log csv for some of this, but we have all the info we need in the log to recover it
PATH = 'data/runs/run6/out.txt'
RESULT_PATH = 'data/runs/run6/1200-test-log.csv'
EPOCH_COUNT = 200
```

```python
with open(PATH, 'r') as f:
    lines = [l.replace('\n', '') for l in f.readlines()]

lines
```

```python
best_train_fit = []
val_fit = []
pop_avg_fit = [0 for i in range(EPOCH_COUNT)] # not available in log

for l in lines:
    if l.startswith('Best training'):
        best_train_fit.append(float(l.split(': ')[1])/100)
    elif l.startswith('Validation set'):
        val_fit.append(float(l.split(': ')[1]))
```

```python
# write to csv
with open(RESULT_PATH, 'w') as f:
    for i in range(EPOCH_COUNT):
        f.write(f'{best_train_fit[i]},{val_fit[i]},{pop_avg_fit[i]}\n')
```
